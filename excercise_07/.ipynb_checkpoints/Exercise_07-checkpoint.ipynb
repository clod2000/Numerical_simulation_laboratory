In this notebook we perform the simulation of a system with $N=108$ particles in a fixed volume with $PBC$ in the canonical ensemble ($NVT$) using $ M(RT)^2$ in the following thermodynamic condition:
- solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
- liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
- gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)
    
### <font color="green"> Exercise 7.2
    
The transition probability of our Metropolis algorithm is described by a uniform distribution between $[-\delta, \delta ]$ so, in order to get a Metropolis acceptance ratio fixed near $0.5$, we need to set the $\delta$ parameter equal to 0.12 for solid, 0.2 for liquid and 0.3 for gas.

After this balancing we proceed calculating the autocorrelation function of the data of the potential energy per particle U/N.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

epot_solid  = np.loadtxt("solid_epot_eq.dat", usecols=(1), unpack = 'true')
epot_liquid = np.loadtxt("liquid_epot_eq.dat", usecols=(1), unpack = 'true')
epot_gas    = np.loadtxt("gas_epot_eq.dat", usecols=(1), unpack = 'true')
  
nsteps = 500
t = np.arange(nsteps)
tmax =50*nsteps  #len(epot_solid) 

def autocorrelation(epot_state) :
    epot = epot_state
    s = [np.zeros(nsteps) for j in range(3) ]
    s[0] = [sum(epot[0:tmax -i]*epot[i:tmax ]) for i in  t] 
    s[1] = [sum(epot[0:tmax -i ] ) for i in  t]
    s[2] = [sum(epot[i:tmax ] ) for i in  t]
    for i in range(3):
        s[i]/=tmax -t
    den = 1./tmax*sum( epot[0:tmax]**2 ) - (1./tmax*sum( epot[0:tmax] ))**2
    return (s[0] -s[1]*s[2])/den
```

```python
Xs = autocorrelation(epot_solid)
Xl = autocorrelation(epot_liquid)
Xg = autocorrelation(epot_gas)
```

```python
fig = plt.figure(figsize = (18,5) )

plt.plot(t,Xs, label = "solid autocorrelation")
plt.plot(t,Xl, label = "liquid autocorrelation")
plt.plot(t,Xg, label = "gas autocorrelation")
plt.ylabel("autocorrelation function")
plt.xlabel("# of data")
plt.grid(True)

plt.legend(loc = 'best')

plt.show()
```

As we can see from the graph above, for all states the correlation between the data decreases rapidly until it is almost zero already after 100 data.

We can then proceed to estimate the uncertainty of the expectation value of U/N for blocks of different size in the data blocking thecnique from $L=M/N=10$ to $L=5\times 10^3$, being $N$ the number of blocks. 


```python
def blk_err(data,L):   #n , block's lenght
    M = len(data)     #num of MC steps
    N = int(M/L)      #num of blocks
    glob_av = 0
    glob_av2 = 0
    
    for i in range(N):
        blk_av = 0
       
        for j in range(L):
            blk_av += data[j+ L*i]
        blk_av /= L
        glob_av += blk_av
        glob_av2 += (blk_av*blk_av)
    glob_av /= N
    glob_av2 /= N
   
    return np.sqrt( (glob_av2 - glob_av**2)/N)

err_solid  = [blk_err(epot_solid,L) for L in range(10,5000,10) ]
err_liquid = [blk_err(epot_liquid,L) for L in range(10,5000,10) ]
err_gas = [blk_err(epot_gas,L) for L in range(10,5000,10) ]
```

```python
L=np.arange(10,5000,10)

fig = plt.figure( figsize = (15,20) )
plt.subplot(3,1,1)
plt.title("Solid")
plt.xlabel("# block size")
plt.ylabel("variance")
plt.grid(True)
plt.plot(L,err_solid, "-*", label = "solid" )

plt.subplot(3,1,2)
plt.title("Liquid")
plt.xlabel("# block size")
plt.ylabel("variance")
plt.plot(L,err_liquid, "-*", color = 'orange' ,label = "liquid" )
plt.grid(True)

plt.subplot(3,1,3)
plt.title("Gas")
plt.xlabel("# block size")
plt.ylabel("variance")
plt.plot(L,err_gas, "-*", color = 'green' ,label = "liquid" )
plt.grid(True)
```

From the above graphs, it is evident that a too small number of elements per block would result in an underestimation of the uncertainty of the expectation value.

In our simulation, we can safely take blocks of 1000 data items each and be sure not to underestimate the error and not to make statistics on correlated data.

### <font color="green"> Exercise 7.3 & 7.4

Now, after the inclusion of the calculation of the radial distribution function, $g(r)$, as a function of the radial distance $r$ between the particles, with $r \ in \left[ 0; L/2\right]$ within the MD/MC code,
we perform a Monte Carlo NVT simulation through a Lennard-Jones model of Argon (in LJ units) in the thermodynamic condtion described at the head of the notebook. 
In addition, we again run the molecular dynamics simulation in the microcanical ensemble to obtain and compare the radial distribution function $g(r)$


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np



def plt_simulation(state):   #state = solid, liquid or gas
    fig, axs = plt.subplots(1,2, figsize = (18,6) )
    fig.suptitle(str(state)+" simulation", fontsize = (25))

    press, error_press   = np.loadtxt("NVT/"+state+"/output_press.dat", usecols=(2,3), unpack = 'true')
    epot, error_epot     = np.loadtxt("NVT/"+state+"/output_epot.dat", usecols=(2,3), unpack = 'true')
    r_mc,g_mc,error_g_mc = np.loadtxt("NVT/"+state+"/output_g.dat", usecols=(0,1,2), unpack = 'true')
    r_md,g_md,error_g_md = np.loadtxt("NVE/"+state+"/output_g.dat", usecols=(0,1,2), unpack = 'true')
    
    N = len(press)
    x = np.arange(N)
    

    axs[0].errorbar(x,press,yerr=error_press , label = "simulation")
    axs[0].set_title("pression")
    axs[0].set_xlabel("blocks")
    axs[0].set_ylabel("P*")
    axs[0].hlines(press[-1] ,0 ,N, color = "yellow" , label = "last block value") 

    axs[1].errorbar(x,epot,yerr=error_epot)
    axs[1].set_title("potential energy")
    axs[1].set_xlabel("blocks")
    axs[1].set_ylabel("V*")
    axs[1].hlines(epot[-1] ,0 ,N, color = "yellow") 

    fig2 = plt.figure( figsize = (20,10) )
    plt.errorbar(r_mc,g_mc,yerr=error_g_mc, label= ("NVT" ) )
    plt.errorbar(r_md,g_md,yerr=error_g_md, label= ("NVE" ) )
    plt.title("Distribution radial function")
    plt.xlabel("r*")
    plt.ylabel("g(r*)")
    plt.grid(True)
    
    plt.legend(loc = "best")

    for i in range(2):
        axs[i].grid(True)
    fig.legend()
    plt.show()
    return
```

```python
plt_simulation("solid")
```

```python
plt_simulation("liquid")
```

```python
plt_simulation("gas")
```

The results are extremely satisfying: the fact that the simulations in the NVT and NVE ensembles will give the same results are tangible proof of the validity of statistical mechanics.
