### <font color="green"> Exercise 6.1
Simulation of the one-dimensional Ising model for a 50-spin system with nearest neighbor interaction.
    
In the first part of the Notebook we verify that the implementation of the sampling algorithms of Metropolis and Gibbs is correct and it follows that the results have a pattern typical of block-averaged graphs by evolving the system for 50 blocks of $10^5$ steps each. 
    
The initial parameters are:
- T = 2
- J = 1
- h = 0
- $\mu_B=1$ and $k_B=1$
    
During the simulations we see that the system reaches an equilibrium configuration since the first steps: considering that the blocks used are composed of $10^5$ steps, we can neglect an initial equilibrium phase and proceed directly with the simulation.

  
 

```python

import matplotlib
import matplotlib.pyplot as plt
import numpy as np


N=50 # numero di blocchi

x = np.arange(N)

fig, axs = plt.subplots(2,2, figsize = (20,15) )
fig.suptitle("Metropolis Simulation", fontsize = (22))

ene, error_ene    = np.loadtxt("metro/output.ene.0", usecols=(2,3), unpack = 'true')
heat, error_heat  = np.loadtxt("metro/output.heat.0", usecols=(2,3), unpack = 'true')
mag, error_mag    = np.loadtxt("metro/output.mag.0", usecols=(2,3), unpack = 'true')
chi, error_chi    = np.loadtxt("metro/output.chi.0", usecols=(2,3), unpack = 'true')


axs[0,0].errorbar(x,ene,yerr=error_ene , label = "simulation")
axs[0,0].set_title("energy")
axs[0,0].set_xlabel("blocks")
axs[0,0].set_ylabel("H")
axs[0,0].hlines(ene[-1] ,0 ,N, color = "yellow", label = "last block value") 
axs[0,0].grid(True)

axs[0,1].errorbar(x,heat,yerr=error_heat)
axs[0,1].set_title("Heat Capacity")
axs[0,1].set_xlabel("blocks")
axs[0,1].set_ylabel("c")
axs[0,1].hlines(heat[-1] ,0 ,N, color = "yellow")
axs[0,1].grid(True)

axs[1,0].errorbar(x,mag,yerr=error_mag)
axs[1,0].set_title("magnetization")
axs[1,0].set_xlabel("blocks")
axs[1,0].set_ylabel("m")
axs[1,0].hlines(mag[-1] ,0 ,N, color = "yellow") 
axs[1,0].grid(True)

axs[1,1].errorbar(x,chi,yerr=error_chi )
axs[1,1].set_title("magnetic subsceptibility")
axs[1,1].set_xlabel("blocks")
axs[1,1].set_ylabel("chi")
axs[1,1].hlines(chi[-1] ,0 ,N, color = "yellow") 
axs[1,1].grid(True)

fig.legend(fontsize = 13)

plt.show()
```

### Comparation of the simulations with the exact curves

After verifying the correctness of the code, we proceed with the verification of the results obtained with the curves obtained analytically by varying the values of T in a range between 0.5 and 2.

Note: the values in the plots are to be considered per unit of spin.

```python
T = np.linspace(0.5,2.0,num=16)
fig, axs = plt.subplots(2,2, figsize = (20,18) )

ene, error_ene    = np.loadtxt("metro/ResEn.dat", usecols=(2,3), unpack = 'true')
heat, error_heat  = np.loadtxt("metro/ResHeat.dat", usecols=(2,3), unpack = 'true')
mag, error_mag    = np.loadtxt("metro/ResMag.dat", usecols=(2,3), unpack = 'true')
chi, error_chi    = np.loadtxt("metro/ResChi.dat", usecols=(2,3), unpack = 'true')

eneG, error_eneG    = np.loadtxt("Gibbs/ResEn.dat", usecols=(2,3), unpack = 'true')
heatG, error_heatG  = np.loadtxt("Gibbs/ResHeat.dat", usecols=(2,3), unpack = 'true')
magG, error_magG    = np.loadtxt("Gibbs/ResMag.dat", usecols=(2,3), unpack = 'true')
chiG, error_chiG    = np.loadtxt("Gibbs/ResChi.dat", usecols=(2,3), unpack = 'true')


Te = np.linspace(0.5,2.0,num=100)
b = 1/Te
J = 1.0
Ns = 50
th = np.tanh(J/Te)
thN= th**Ns
ch = 1/th

e = -J*( th + ch*thN )/( 1 + thN )
he=((b*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

h=0.02 #external field
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns

M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
X = b*np.exp(2*b*J)*(1-thN)/(1+thN)

axs[0,0].errorbar(T,ene,yerr=error_ene, fmt='go' , label = 'Metropolis')
axs[0,0].errorbar(T,eneG,yerr=error_eneG, fmt='ro' , label = "Gibbs" , alpha=0.8)
axs[0,0].plot(Te, e, label = "analitic value" )
axs[0,0].set_title("energy (h = 0)",fontsize = 18)
axs[0,0].set_xlabel("T",fontsize = 15)
axs[0,0].set_ylabel("H",fontsize = 15)
 
axs[0,1].errorbar(T,heat,yerr=error_heat, fmt='go' )
axs[0,1].errorbar(T,heatG,yerr=error_heatG, fmt='ro')
axs[0,1].plot(Te, he)
axs[0,1].set_title("Heat Capacity (h = 0)",fontsize = 18)
axs[0,1].set_xlabel("T",fontsize = 15)
axs[0,1].set_ylabel("c",fontsize = 15)

axs[1,0].errorbar(T,mag,yerr=error_mag, fmt='go'  )
axs[1,0].errorbar(T,magG,yerr=error_magG, fmt='ro' )
axs[1,0].plot(Te, M)
axs[1,0].set_title("magnetization (h = 0.02)",fontsize = 18)
axs[1,0].set_xlabel("T",fontsize = 15)
axs[1,0].set_ylabel("m",fontsize = 15)

axs[1,1].errorbar(T,chi,yerr=error_chi, fmt='go' )
axs[1,1].errorbar(T,chiG,yerr=error_chiG, fmt='ro')
axs[1,1].plot(Te, X)
axs[1,1].set_title("magnetic subsceptibility (h = 0)",fontsize = 18)
axs[1,1].set_xlabel("T",fontsize = 15)
axs[1,1].set_ylabel("$\chi$",fontsize = 15) 

fig.legend( fontsize = 15 )

plt.show()
```
