```python

import matplotlib
import matplotlib.pyplot as plt
import numpy as np



fig, axs = plt.subplots(2,2, figsize = (20,18) )
fig.suptitle("Metropolis Equilibration", fontsize = (22))

d,ene, error_ene    = np.loadtxt("ISING_1D/equilibration1/output.ene.0", usecols=(0,2,3), unpack = 'true')
heat, error_heat  = np.loadtxt("ISING_1D/equilibration1/output.heat.0", usecols=(2,3), unpack = 'true')
mag, error_mag    = np.loadtxt("ISING_1D/equilibration1/output.mag.0", usecols=(2,3), unpack = 'true')
chi, error_chi    = np.loadtxt("ISING_1D/equilibration1/output.chi.0", usecols=(2,3), unpack = 'true')

ene2, error_ene2    = np.loadtxt("ISING_1D/equilibration2/output.ene.0", usecols=(2,3), unpack = 'true')
heat2, error_heat2  = np.loadtxt("ISING_1D/equilibration2/output.heat.0", usecols=(2,3), unpack = 'true')
mag2, error_mag2    = np.loadtxt("ISING_1D/equilibration2/output.mag.0", usecols=(2,3), unpack = 'true')
chi2, error_chi2    = np.loadtxt("ISING_1D/equilibration2/output.chi.0", usecols=(2,3), unpack = 'true')
x = np.arange(np.size(d))

axs[0,0].errorbar(x,ene,yerr=error_ene )
axs[0,0].errorbar(x,ene2,yerr=error_ene2 )
axs[0,0].set_title("energy")
axs[0,0].set_xlabel("blocks")
axs[0,0].set_ylabel("H")


axs[0,1].errorbar(x,heat,yerr=error_heat)
axs[0,1].errorbar(x,heat2,yerr=error_heat2)
axs[0,1].set_title("Heat Capacity")
axs[0,1].set_xlabel("blocks")
axs[0,1].set_ylabel("c")
 

axs[1,0].errorbar(x,mag,yerr=error_mag)
axs[1,0].errorbar(x,mag2,yerr=error_mag2)
axs[1,0].set_title("magnetization")
axs[1,0].set_xlabel("blocks")
axs[1,0].set_ylabel("m")


axs[1,1].errorbar(x,chi,yerr=error_chi )
axs[1,1].errorbar(x,chi2,yerr=error_chi2 )
axs[1,1].set_title("magnetic subsceptibility")
axs[1,1].set_xlabel("blocks")
axs[1,1].set_ylabel("chi")
000
```
