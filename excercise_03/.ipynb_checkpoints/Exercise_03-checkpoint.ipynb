### <font color="green"> Exercise 3.1
Simulation via Montecarlo of the "European call-option and put-option price" at time t=0 and comparision between the results and the Black_Scholes analitic solution.

1. Direct sampling of the final asset price $S(T)$ for a geometric Brownian motion (GBM) with constant risk-free interest rate $r$ and volatility $\sigma^2$. 

The paramaters used are:
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

N=100
x = np.arange(N)

#Black-Scholes analitic result:
call = 14.975790778311286
put = 5.4595325819072364

#call direct
call_dir,c_d_error = np.loadtxt("Call_direct.dat", usecols=(0,1),delimiter=' ', unpack='true')

plt.errorbar(x,call_dir,yerr=c_d_error, label = "simulation")
plt.xlabel('# blocks')
plt.ylabel('Call_option')
plt.hlines(call,0,100,color="red", label = "theoretical value")
plt.legend()
plt.grid(True)
plt.show()

call_mean = round(call_dir[-1],2)
call_sigma = round(c_d_error[-1],2)

print("Call_option: ",call_mean, "+/- ", call_sigma)
print("confidence with theoretical value: ",round(np.abs((call-call_mean)/call_sigma ),2),"sigma")

#put direct
put_dir,p_d_error = np.loadtxt("Put_direct.dat", usecols=(0,1),delimiter=' ', unpack='true')

plt.errorbar(x,put_dir,yerr=p_d_error, label = "simulation")
plt.xlabel('# blocks')
plt.ylabel('Put_option')
plt.hlines(put,0,100,color="green", label = "theoretical value")
plt.legend()
plt.grid(True)
plt.show()

put_mean = round(put_dir[-1],2)
put_sigma = round(p_d_error[-1],2)

print("Put_option: ",put_mean, "+/- ", put_sigma)
print("confidence with theoretical value: ",round(np.abs((put-put_mean)/put_sigma ),2),'sigma')

```

2. Sampling of the discretized $GBM(r,\sigma^2)$ path of the asset price S dividing $[0,T]$ in $100$ time intervals: $S(0)$, $S(t_1)$, $S(t_2)$, ... up to $S(t_{100}=T=1)$

```python
N=100
x = np.arange(N)

#Black-Scholes analitic result:
call = 14.975790778311286
put = 5.4595325819072364

#call direct
call_discr,c_error = np.loadtxt("Call_discr.dat", usecols=(0,1),delimiter=' ', unpack='true')

plt.errorbar(x,call_discr,yerr=c_error, label = "simulation")
plt.xlabel('# blocks')
plt.ylabel('Call_option')
plt.hlines(call,0,100,color="red", label = "theoretical value")
plt.legend()
plt.grid(True)
plt.show()

call_mean = round(call_discr[-1],2)
call_sigma = round(c_error[-1],2)

print("Call_option: ",call_mean, "+/- ", call_sigma)
print("confidence with theoretical value: ",round(np.abs((call-call_mean)/call_sigma ),2),"sigma")

#put direct
put_discr,p_error = np.loadtxt("Put_discr.dat", usecols=(0,1),delimiter=' ', unpack='true')

plt.errorbar(x,put_discr,yerr=p_error, label = "simulation")
plt.xlabel('# blocks')
plt.ylabel('Put_option')
plt.hlines(put,0,100,color="green", label = "theoretical value")
plt.legend()
plt.grid(True)
plt.show()

put_mean = round(put_discr[-1],2)
put_sigma = round(p_error[-1],2)

print("Put_option: ",put_mean, "+/-", put_sigma)
print("confidence with theoretical value: ",round(np.abs((put-put_mean)/put_sigma ),2),"sigma")
```
