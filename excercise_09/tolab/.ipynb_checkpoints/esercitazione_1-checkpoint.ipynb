### <font color="red"> excercise 1.1   
Test of the pseudo-random number generator downloaded from the NSL Ariel web site:
    
1. Estimate of the value of the integral $\langle r \rangle = \int_0^1 r dr $ and graphic representation of $\langle r \rangle - 1/2  $ and respective error as the number of throws increases. (as function of the number of block N)
 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M=100000
N=100
L=int(M/N)
x = np.arange(N) 

x*=L
sum_prog,error = np.loadtxt("ex_1.1.1.dat", usecols=(0,1),delimiter=' ', unpack='true')

plt.errorbar(x,sum_prog -0.5,yerr=error, label="simulazione")
plt.hlines(0,0,M,color="red", label = "valore teorico")
plt.legend()
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)

plt.show()

print("Miglior stima dell'integrale: ", round(sum_prog[-1],3), " +/- ", round(error[-1],3)  )
print("confidenza col valore teorico: ",round ((sum_prog[-1] - 0.5)/error[-1],1),"sigma")
```

2. Calcolo della deviazione standard $\sigma^2 = \int_0^1 (r-1/2)^2 dr$ e rappresentazione grafica di $\langle \sigma^2 \rangle -1/12$ con rispettive barre di errore al crescere del numero M di estrazioni espresse in funzione del numero di blocchi N.

```python
sum_prog,error = np.loadtxt("ex_1.1.2.dat", usecols=(0,1),delimiter=' ', unpack='true')
plt.errorbar(x,sum_prog - 1/12,yerr=error)
plt.errorbar(x,sum_prog -1/12,yerr=error, label="simulazione")
plt.hlines(0,0,M,color="red", label = "valore teorico")
plt.legend()
plt.xlabel('#throws')
plt.ylabel('<(r - 0,5)^2> - 1/12')
plt.grid(True)

plt.show()
print("Miglior stima dell'integrale: ", round(sum_prog[-1],4)," +/- ",round(error[-1],4) )
print("confidenza col valore teorico: ",round ((sum_prog[-1] - 1/12)/error[-1],3),"sigma")
```

 ### <font color="red">Esercizio 1.2
Verifica del teorema del limite centrale.
Di seguito vengono riportati tre figure contenenti gli istogrammi di $10^4$ realizzazioni di $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ (per $N=1, 2, 10, 100$), rispettivamente per le distribuzioni uniforme, eponenziale e lorentziana. 
Per l'istogramma con $N=100$ si ha eseguito un fit con una gaussiana per le prime due distribuzioni e con una Cauchy-Lorentz per la terza.

```python
from scipy.stats import norm

N=[1,2,10,100]

x=np.arange(10000)

sum= np.loadtxt('unif_1.2.dat', usecols=(0,1,2,3), delimiter=' ', unpack='true')

n_bins = 100
fig, axs = plt.subplots(2,2, figsize = (16,16) )
axs[0,0].hist(sum[0], n_bins, density= 'true')
axs[0,1].hist(sum[1], n_bins, density= 'true')
axs[1,0].hist(sum[2], n_bins, density= 'true')
axs[1,1].hist(sum[3], n_bins, density= 'true')

mean,std=norm.fit(sum[3])
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
y = norm.pdf(x, mean, std)
axs[1,1].plot(x,y)

for i in range(2) :
    for j in range(2):
        axs[i,j].set_xlabel('range')
        axs[i,j].set_ylabel('frequency')
        axs[i,j].grid(True)
```

### <font color="red">Esercizio 1.3
Simulazione dell'esperimento di Buffon per ricavare il valore di $\pi$.
Un bastone di lunghezza $L$ viene lanciato casualmente lungo un piano con delle righe orizzontali a distanza $d$.
La probabilità di intersezione del bastone con una delle linee è  $P = 2L/\pi d$. Effettuando $N$ lanci e contando quante volte il bastoncino colpisce una riga ($N_{hit}$) è possibile stimare il valore di $\pi$ secondo la seguente relazione:
    $$\pi = \frac{2L}{Pd} = \lim_{N\to \infty}\frac{2LN}{N_{hit}d}$$  
    
Il grafico seguente viene realizzato 
    
<span style="color:RED"> DA SISTEMARE INTESTAZIONE e aggiungere valore atteso ( e modificare cambionamento angolo in cpp) </span>
  

```python
M=100000
N=100
K=int(M/N)
x = np.arange(N) 

pi_prog, error = np.loadtxt("pi.dat", usecols=(1,2), delimiter=' ', unpack='true')

plt.errorbar(x,pi_prog ,yerr=error, label="simulation")
plt.xlabel('#blocks')
plt.ylabel('$\pi$')
plt.hlines(np.pi,0,N, color = "yellow", label = "theoretical value")
plt.legend()
plt.grid(True)
plt.show()

mean,sigma = np.loadtxt("pi_final.dat", usecols=(0,1), delimiter=' ')
print("valore finale pi_greco: " , round(mean,3) , " +\- " ,round(sigma,3))
```
