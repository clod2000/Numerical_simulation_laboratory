### <font color="green"> Molecular dynamic in the Microcanonical Ensemble (NVE)
In this Notebook we want to perform a Molecular dynamics simulation via Lennard-Jones of Argon in solid,liquid and gas phases using the program downloaded from the NSL ariel site.
       
The program perform the computation of the total (internal) energy per particle, $E/N$, the potential energy per particle, $U/N$, the kinetic energy per particle, $K/N$ and the temperature, $T$.
After adding the estimation of pressure, $P$, we proceeded with the simulation in different thermodynamic states.
    
Working on a NVE ensemble, we expect that the energy per particle, $E/N$, will be conserved.
    
Note: the variables are used in reduced unit.


# Solid phase
Initial conditions:

- $\rho$* $=1.1$
- $ùëá$* $=0.8$
- cut-off radius: $r_c = 2.2$

The first step consists in the equilibration of the temperature:
the simulation starts from a perfect crystal configuration until evolving to the configuration of the molecules in the desired phase.
This involves an increase of potential energy (because the crystal configuration is the minimum of Lennard-jones potential), consequently a decrease of kinetic energy and therefore of temperature until it stabilizes around a certain value.
In order for that specific value to be T*, we have made a series of attempts starting from a higher input temperature.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize = (18,5) )
fig.suptitle("Equilibration phase", fontsize = (22))

N,temp_ist = np.loadtxt("solido/solid_equilibration.dat", usecols=(0,1), unpack = 'true')
x = np.arange(N[-1])
 
plt.plot(x,temp_ist, label = "misured temperature")
plt.title("temperature")
plt.xlabel("steps")
plt.ylabel("T*")
plt.hlines(0.8 ,0 ,N[-1], color = "red", label = "desired value") 

plt.legend()

plt.show()
```

We found the results in the previous graph by starting the simulation at the temperature T* = 1.541.


Always from the previous graph we see that if we move the system for at least 300 steps before starting the simulation, the temperature T* fluctuates near the desired value.

In the following part we perform the simulation with 100 blocks from 3000 step each, after evolving the system to the equilibrium.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


N=100 # number of blocks

x = np.arange(N)

fig, axs = plt.subplots(3,2, figsize = (15,15) )
fig.suptitle("Simulation", fontsize = (32))

press, error_press = np.loadtxt("solido/output_press.dat", usecols=(2,3), unpack = 'true')
ekin, error_ekin   = np.loadtxt("solido/output_ekin.dat", usecols=(2,3), unpack = 'true')
epot, error_epot   = np.loadtxt("solido/output_epot.dat", usecols=(2,3), unpack = 'true')
etot, error_etot   = np.loadtxt("solido/output_etot.dat", usecols=(2,3), unpack = 'true')
temp, error_temp   = np.loadtxt("solido/output_temp.dat", usecols=(2,3), unpack = 'true')

plt.subplot(3,2,1)
plt.errorbar(x,press,yerr=error_press , label = "simulation")
plt.title("pression")
plt.xlabel("blocks")
plt.ylabel("P*")
plt.hlines(press[-1] ,0 ,N, color = "yellow" , label = "last block value") 
plt.grid(True)

plt.subplot(3,2,2)
plt.errorbar(x,ekin,yerr=error_ekin)
plt.title("kinetic energy")
plt.xlabel("blocks")
plt.ylabel("K*")
plt.hlines(ekin[-1] ,0 ,N, color = "yellow")
plt.grid(True)

plt.subplot(3,2,3)
plt.errorbar(x,epot,yerr=error_epot)
plt.title("potential energy")
plt.xlabel("blocks")
plt.ylabel("V*")
plt.hlines(epot[-1] ,0 ,N, color = "yellow") 
plt.grid(True)

plt.subplot(3,2,4)
plt.errorbar(x,etot,yerr=error_etot )
plt.title("total energy")
plt.xlabel("blocks")
plt.ylabel("E*")
plt.hlines(etot[-1] ,0 ,N, color = "yellow") 
plt.grid(True)

plt.subplot(3,2,5)
plt.errorbar(x,temp,yerr=error_temp )
plt.title("temperature")
plt.xlabel("blocks")
plt.ylabel("T*")
plt.hlines(temp[-1] ,0 ,N, color = "yellow") 
plt.hlines(0.8 ,0 ,N, color = "red" , label = "desired temperature") 
plt.grid(True)

plt.subplot(3,2,6)
plt.axis("off")

fig.legend( fontsize = 15)

plt.show()
```

# Liquid phase

Initial conditions:

- $\rho$* $=0.8$
- $ùëá$* $=1.1$
- cut-off radius: $r_c = 2.5$

The equilibrium phase of the liquid proceeds similarly to that of the solid: again, the starting configuration is very close to the potential minimum, so we expect to have to start at a temperature above that of interest to reach equilibrium.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize = (18,5) )
fig.suptitle("Equilibration phase", fontsize = (22))

N,temp_ist = np.loadtxt("liquido/liquid_equilibration.dat", usecols=(0,1), unpack = 'true')
x = np.arange(N[-1])
 
plt.plot(x,temp_ist, label = "misured temperature")
plt.title("temperature")
plt.xlabel("steps")
plt.ylabel("T*")
plt.hlines(1.1 ,0 ,N[-1], color = "red", label = "desired value") 

plt.legend(fontsize = 14)
plt.show()
```

The equilibration for the liquid is reached using T*=1.983.

In the following part we perform the simulation with 100 blocks from 3000 step each, after evolving the system to the equilibrium moving it for 5000 steps.

# Gas phase

Initial conditions:

- $\rho$* $=0.05$
- $ùëá$* $=1.2$
- cut-off radius: $r_c = 5.0$


Gas balancing is a bit more complicated:
since the density to be simulated is much smaller, the size of the particles for each box in the simulation is also smaller and consequently their interaction is rarer . This means that we can find ourselves rather far from the minimum of the Lennard-Jones potential and that the system evolves towards a lower potential (with consequent increase of the kinetic energy and therefore of the temperature).
The equilibrium temperature is therefore obtained starting from lower test values.
The following plot also shows how the value of T* continues to fluctuate around the value we want to obtain and we recognize a "scale trend" that repeats more or less periodically.
In order for our results to make sense, we took 20 blocks of 20000 steps each, so as to mediate these "scaling trends".

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize = (16,10) )
fig.suptitle("Equilibration phase", fontsize = (22))

N,temp_ist = np.loadtxt("gas/gas_equilibration.dat", usecols=(0,1), unpack = 'true')
x = np.arange(N[-1])

plt.plot(x,temp_ist, label = "misured temperature")
plt.title("temperature")
plt.xlabel("steps")
plt.ylabel("T*")
plt.hlines(1.2 ,0 ,N[-1], color = "red", label = "desired value") 

plt.legend()

plt.show()
```

The equilibration for the gas is reached using T*=1.01.

In the following part we perform the simulation with 20 blocks from 20000 step each, after evolving the system to the equilibrium moving it for 40000 steps.
